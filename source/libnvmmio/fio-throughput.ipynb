```python
import matplotlib.pyplot as plt, numpy as np, random as rd, pandas as pd
from matplotlib import cm
%matplotlib inline
```

```python
fig, (nvdimms, optane) = plt.subplots(ncols=2, figsize=(10, 3))
workloads = ['SR', 'RR', 'SW', 'RW']
xpos = np.arange(len(workloads))
width = 0.2
#mycm = cm.get_cmap('cividis', 3)
mycm = cm.get_cmap('viridis', 3)

ext4 = [3.3604,2.4173,2.0227,1.6767]
pmfs = [2.920, 2.447, 1.674, 1.552]
nova = [3.863, 2.549, 1.893, 1.615]
lnvm = [6.676, 5.824, 3.365, 3.033]

nvdimms.bar(xpos-0.3, ext4, width, capsize=5, label='Ext4-DAX', zorder=2, edgecolor='black', color=mycm.colors[0])
nvdimms.bar(xpos-0.1, pmfs, width, capsize=5, label='PMFS', zorder=2, edgecolor='black', color=mycm.colors[1])
nvdimms.bar(xpos+0.1, nova, width, capsize=5, label='NOVA', zorder=2, edgecolor='black', color=mycm.colors[2])
nvdimms.bar(xpos+0.3, lnvm, width, capsize=5, label='Libnvmmio', zorder=2, edgecolor='black', color=mycm.colors[2], hatch="xx")
nvdimms.grid(linestyle='--', linewidth='0.5', zorder=0)
nvdimms.tick_params(axis='both', which='major', labelsize=14)

nvdimms.set_ylabel('Bandwidth (GiB/s)', fontsize=16)
nvdimms.set_xlabel('Access Pattern', fontsize=16)
nvdimms.set_title('(A) NVDIMMs', fontsize=18)
nvdimms.set_xticks(xpos)
nvdimms.set_xticklabels(workloads)

ext4 = [1.916,1.277,1.134,1.012]
pmfs = [1.427,1.121,1.259,0.992]
nova = [1.773,1.223,1.111,1.059]
lnvm = [2.486,2.232,1.437,1.268]

optane.bar(xpos-0.3, ext4, width, capsize=5, label='Ext4-DAX', zorder=2, edgecolor='black', color=mycm.colors[0])
optane.bar(xpos-0.1, pmfs, width, capsize=5, label='PMFS', zorder=2, edgecolor='black', color=mycm.colors[1])
optane.bar(xpos+0.1, nova, width, capsize=5, label='NOVA', zorder=2, edgecolor='black', color=mycm.colors[2])
optane.bar(xpos+0.3, lnvm, width, capsize=5, label='Libnvmmio', zorder=2, edgecolor='black', color=mycm.colors[2], hatch="xx")
optane.grid(linestyle='--', linewidth='0.5', zorder=0)
optane.tick_params(axis='both', which='major', labelsize=14)
optane.set_xlabel('Access Pattern', fontsize=16)
optane.set_title('(B) Optane', fontsize=18)
optane.set_xticks(xpos)
optane.set_xticklabels(workloads)
optane.set_ylim(0, 4.2)
optane.legend(ncol=2, fontsize=14, handlelength=1.0)

optane.axhline(y=2.5, xmin=0.03, xmax=0.51, color='r', linewidth=2, linestyle='--')
optane.axhline(y=1.46, xmin=0.52, xmax=0.97, color='r', linewidth=2, linestyle='--')

fig.savefig('fio-throughput.pdf', bbox_inches='tight', pad_inches=0)
```
